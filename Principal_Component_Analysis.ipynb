```python
np.random.seed(40)
#GNB
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import RandomizedSearchCV
gnb_score  =[]
gnbt_score=[]
gnb_pre=[]
gnb_rec=[]
gnb_f1=[]
lr_sen=[]
lr_roc=[]
for train_index, test_index in kf.split(X):
    X_train, y_train = X[train_index], y[train_index]
    X_test, y_test   = X[test_index], y[test_index]
    
    parameters = {'var_smoothing': (1e-09,1e-08,1e-07,1e-06,1e-05,1e-04,1e-03,1e-02,1e-01,1,10,5e-09,5e-08,5e-07,5e-06,5e-05,
                                5e-04,5e-03,5e-02,5e-01,5)}
    DT_grid  = RandomizedSearchCV( GaussianNB(), param_distributions = parameters, cv = 5, verbose = True,scoring='accuracy')
    result=DT_grid.fit(X_train,y_train)
    
    
    gnb = result.best_estimator_
    pred_gnb = gnb.predict(X_test)
    y_pred_proba = gnb.predict_proba(X_test)
    accur_gnb=accuracy_score(y_test,pred_gnb)
    pred1_gnb=gnb.predict(X_train)
    accur1_gnb=accuracy_score(y_train,pred1_gnb)
    gnb_score.append(accur_gnb)
    gnbt_score.append(accur1_gnb)
    prec_gnb=precision_score(y_test,pred_gnb,average='weighted')
    gnb_pre.append(prec_gnb)
    rec_gnb=recall_score(y_test,pred_gnb,average='weighted')
    gnb_rec.append(rec_gnb)
    f1_gnb=f1_score(y_test,pred_gnb,average='weighted')
    gnb_f1.append(f1_gnb)
    gnb_confusion=confusion_matrix(y_test,pred_gnb)
    print(gnb_confusion)
    sensitivity = gnb_confusion[0,0]/(gnb_confusion[0,0]+gnb_confusion[0,1])
    ROC_score=roc_auc_score(y_test,y_pred_proba , average='weighted', multi_class='ovr')
    lr_sen.append(sensitivity)
    lr_roc.append(ROC_score)
    print(gnb)
print('ROC_Score : ',np.mean (lr_roc))
print('Sensitivity : ',np.mean (lr_sen))
print('Scores for gnb :',gnb_score)
print('Mean Test Score for gnb:',np.mean(gnb_score))
print('Scores for gnbt :',gnbt_score)
print('Mean Training Score for gnbt:',np.mean(gnbt_score))
print('Precision Scores for gnb :',gnb_pre)
print('Mean Precision Score for gnb:',np.mean(gnb_pre))
print('Recall Scores for gnb :',gnb_rec)
print('Mean Recall Score for gnb:',np.mean(gnb_rec))
print('f1 Scores for gnb :',gnb_f1)
print('Mean f1 Score for gnb:',np.mean(gnb_f1))
```
